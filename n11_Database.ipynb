```python
def ürün_bilgileri_bul(url):
    try:
        global ekran_boyutu,disk_türü,işlemci,disk_kapasite,işletim_sistemi,model,ram,işlemci_nesli,model_no
        istek2=requests.get(url,headers=header,verify=False)
        soup2=BeautifulSoup(istek2.content,"lxml")
        ürün_ad=soup2.find("h1",attrs={"class":"proName"}).get_text().strip().upper()
        ürün_resim=soup2.find("div",attrs={"class":"imgObj"}).a.get("href")
        ürün_fiyat=soup2.find("div",attrs={"class":"unf-p-summary-price"}).get_text().strip()
        global ürün_puan
        try:
            ürün_puan=soup2.find("div",attrs={"class":"ratingCont"}).find("strong").get_text().strip()
        except :
            ürün_puan=0
        ürün_aciklamalar=soup2.find_all("li",attrs={"class":"unf-prop-list-item"})
        uzunluk=len(ürün_aciklamalar)
        ürün_marka=ürün_aciklamalar[uzunluk-1].find("p",attrs={"class":"unf-prop-list-prop"}).get_text().strip().upper()
        print(ürün_marka)
        aciklamalar=soup2.find_all("li",attrs={"class":"unf-prop-list-item"})
        for i in aciklamalar:
            title=i.find("p",attrs={"class":"unf-prop-list-title"}).get_text().strip().replace(" ","")
            if title.find("EkranBoyutu")!=-1:
                ekran_boyutu=i.find("p",attrs={"class":"unf-prop-list-prop"}).get_text().replace("\"","").strip()
            elif title.find("DiskTürü")!=-1:
                disk_türü=i.find("p",attrs={"class":"unf-prop-list-prop"}).get_text().strip()
            elif title.find("BellekKapasitesi")!=-1:
                ram=i.find("p",attrs={"class":"unf-prop-list-prop"}).get_text().strip()
            elif title.find("İşlemciModeli")!=-1:
                işlemci_nesli=i.find("p",attrs={"class":"unf-prop-list-prop"}).get_text().strip()
                if işlemci_nesli.find("Intel")!=-1:
                    bul=işlemci_nesli.find("-")
                    if bul!=-1:
                        if işlemci_nesli[bul+1].find("1")!=-1:
                            işlemci_nesli=işlemci_nesli[bul+1:bul+3]+".NESIL"
                        else:
                             işlemci_nesli=işlemci_nesli[bul+1:bul+2]+".NESIL"
                işlemci_nesli=işlemci_nesli.upper()
            elif title.find("İşlemci")!=-1 and len("İşlemci")==len(title):
                print("")
                işlemci=i.find("p",attrs={"class":"unf-prop-list-prop"}).get_text().strip().upper()
            elif title.find("DiskKapasitesi")!=-1:
                disk_kapasite=i.find("p",attrs={"class":"unf-prop-list-prop"}).get_text().strip().upper()
            elif title.find("İşletimSistemi")!=-1:
                işletim_sistemi=i.find("p",attrs={"class":"unf-prop-list-prop"}).get_text().strip().upper().replace(" ","")
                if işletim_sistemi.find("WINDOWS")!=-1:
                    işletim_sistemi="WINDOWS"
                elif işletim_sistemi.find("YOK")!=-1:
                    işletim_sistemi="FREEDOS"
                elif işletim_sistemi.find("DOS")!=-1:
                    işletim_sistemi="FREEDOS"
                elif işletim_sistemi.find("CHROMEOS")!=-1:
                        işletim_sistemi="CHROMEOS"
                elif işletim_sistemi.find("MACOS")!=-1:
                        işletim_sistemi="MACOS"
            elif title.find("Model")!=-1 and  title.find("Ekran Kartı Modeli")==-1:
                model=i.find("p",attrs={"class":"unf-prop-list-prop"}).get_text().strip().upper()
            
    except:
        pass
    else:
        print(ürün_marka,ürün_ad,model,işletim_sistemi,işlemci,işlemci_nesli,ram,disk_kapasite,disk_türü,ekran_boyutu,ürün_puan,ürün_resim,ürün_fiyat,url)
        ürün_bilgileri.append([ürün_marka,ürün_ad,model,işletim_sistemi,işlemci,işlemci_nesli,ram,disk_kapasite,disk_türü,ekran_boyutu,ürün_puan,ürün_resim,ürün_fiyat,url])

site_url="https://www.n11.com/bilgisayar/dizustu-bilgisayar"
ürün_bilgileri=[]
for k in range(2,50):
    try:
        istek=requests.get(site_url,headers=header,verify=False)
        soup=BeautifulSoup(istek.content,"lxml")
        gt1=soup.find_all("div",attrs={"class":"pro"})
        ürün_url=[]
        for i in gt1:
            ürün_url.append(i.a.get("href"))
        for i in ürün_url:
            ürün_bilgileri_bul(i)
        site_url="https://www.n11.com/bilgisayar/dizustu-bilgisayar?pg="+str(k)
    except:
        pass
  
```

```python
import pypyodbc 
import datetime
import time
db = pypyodbc.connect(
    'Driver={SQL Server};'
    'Server=.;'
    'Database=UrunlerDeneme;'
    'Trusted_Connection=True;'
)
sqlimlec=db.cursor()
global siteId
def sonMarkaId():
    kontrol=sqlimlec.execute('SELECT * FROM tbl_markalar ORDER BY MarkaId DESC')
    tablo = sqlimlec.fetchone()
    return tablo[0]
def marka_ekle(markaAd):
    sqlimlec.execute('SELECT * FROM tbl_markalar')
    tablo = sqlimlec.fetchall()
    kontrol1=markaAd.replace(" ","")
    if tablo:
        for i in range(0,len(tablo)):
            kontrol2=tablo[i][1].replace(" ","")
            if kontrol2.find(kontrol1)!=-1 or kontrol1.find(kontrol2)!=-1:
                return tablo[i][0]
    veriler=[]
    komut = 'INSERT INTO tbl_markalar VALUES(?)'
    veriler.append(markaAd)
    sonuc = sqlimlec.execute(komut,veriler)
    db.commit()
    markaId=sonMarkaId()
    return markaId
def model_id():
    kontrol=sqlimlec.execute('SELECT * FROM tbl_modeller ORDER BY ModelId DESC')
    tablo = sqlimlec.fetchone()
    return tablo[0]
def model_ekle(markaId,modelNo,modelResim,modelPuan):
    komut = 'INSERT INTO tbl_modeller VALUES(?,?,?,?)'
    veriler=(modelNo,markaId,modelResim,modelPuan)
    sonuc = sqlimlec.execute(komut,veriler)
    db.commit()
    mdl=model_id()
    return mdl

def model_kontrol(modelNo,markaId):
    sqlimlec.execute('SELECT * FROM tbl_modeller')
    tablo = sqlimlec.fetchall()
    kontrol1=modelNo.replace(" ","")
    if tablo:
        for i in tablo:
            kontrol2=i[1].replace(" ","")
            if (kontrol2.find(kontrol1)!=-1 or kontrol1.find(kontrol2)!=-1) and i[2]==markaId:
                return  i[0]
        
    return -1
def urun_ekle(ürünAd,modelId,fiyat,url):
                komut1 = 'INSERT INTO tbl_urunler VALUES(?,?,?,?,?)'
                tarih =  datetime.datetime.now()
                veriler=(ürünAd,modelId,url,fiyat,tarih)
                sqlimlec.execute(komut1,veriler)
                db.commit()
                urn=sonurunId_bul()
                return urn
def sonurunId_bul():
    sqlimlec.execute('SELECT * FROM tbl_urunler ORDER BY UrunId DESC')
    tablo = sqlimlec.fetchone()
    return tablo[0]
def urun_aciklama_ekle(urunId,markaId,isletimSistemi,islemciTip,islemciNesli,ram,diskBoyut,diskTur,ekran_boyutu):
    komut = 'INSERT INTO tbl_urunAciklama VALUES(?,?,?,?,?,?,?,?,?,?)'
    veriler=(urunId,markaId,isletimSistemi,islemciTip,islemciNesli,ram,diskBoyut,diskTur,ekran_boyutu,siteId)
    sqlimlec.execute(komut,veriler)
    db.commit()
def urun_guncelle(fiyat,Url,UrunId):
               sonuc =sqlimlec.execute('UPDATE tbl_urunler SET Fiyat = ? ,UrunUrl= ? WHERE UrunId = ?',(fiyat,Url,UrunId))
               db.commit()
def ayni_urun_varmi(islemciTip,islemciNesli,ram,diskBoyut,diskTur,isletimSistemi,ekran_boyutu):
    komut='SELECT * FROM tbl_urunAciklama  Where isletimSistemi=? and islemciNesli=? and ram=? and diskBoyut=? and diskTur=? and islemciTip=? and ekranBoyut=? and siteId=?'
    veriler=(isletimSistemi,islemciNesli,ram,diskBoyut,diskTur,islemciTip,ekran_boyutu,siteId)
    sqlimlec.execute(komut,veriler)
    bilgiler=sqlimlec.fetchone()
    if bilgiler:
        return bilgiler[1]
    return -1
def ayni_site_model(modelId):
    komut='SELECT t2.UrunId FROM tbl_urunAciklama t1,tbl_urunler t2 where t1.UrunId=t2.UrunId and t2.modelId=? and t1.siteId=?'
    veriler=(modelId,siteId)
    sqlimlec.execute(komut,veriler)
    bilgiler=sqlimlec.fetchone()
    if bilgiler:
        return bilgiler
    return -1
def tablo_bosmu(tabloAd):
    sqlimlec.execute('SELECT * FROM '+tabloAd)
    tablo = sqlimlec.fetchall()
    return tablo
def sontSiteId_bul():
    sqlimlec.execute('SELECT * FROM tbl_siteler ORDER BY SiteId DESC')
    tablo = sqlimlec.fetchone()
    return tablo[0]
def site_ekle(siteIsmi,siteLogo):
    sqlimlec.execute('SELECT * FROM tbl_siteler')
    tablo = sqlimlec.fetchall()
    if tablo:
        for i in range(0,len(tablo)):
            if tablo[i][1].find(siteIsmi)!=-1:
                return tablo[i][0]
    veriler=[]
    komut = 'INSERT INTO tbl_siteler VALUES(?,?)'
    veriler=(siteIsmi,siteLogo)
    sonuc = sqlimlec.execute(komut,veriler)
    db.commit()
    site=sontSiteId_bul()
    return site


siteId=site_ekle("n11","https://www.changersla.co/wp-content/uploads/n11-logo.jpg")
print(len(ürün_bilgileri))
for x in ürün_bilgileri:
    ürünMarka=x[0]
    ürünAd=x[1]
    ModelNo=x[2]
    isletimSistemi=x[3]
    islemciTip=x[4]
    islemciNesli=x[5]
    ram=x[6]
    diskBoyut=x[7]
    diskTur=x[8]
    ekranBoyutu=x[9]
    ürünPuan=float(x[10].replace(",","."))
    ürünResim=x[11]
    fiyat=float(x[12].replace(".","").replace(",","."))
    url=x[13]
    kontrol=-1
    markaId=marka_ekle(ürünMarka)
    kontrol=model_kontrol(ModelNo,markaId)
    if kontrol!=-1:
        modelId=kontrol
        bilgiler=ayni_urun_varmi(islemciTip,islemciNesli,ram,diskBoyut,diskTur,isletimSistemi,ekranBoyutu)
        if bilgiler!=-1:
            UrunId=bilgiler
            urun_guncelle(fiyat,url, UrunId)
        else:
            ayniId=ayni_site_model(modelId)
            if ayniId==-1:
                UrunId=urun_ekle(ürünAd,modelId,fiyat,url)
                urun_aciklama_ekle(UrunId,markaId,isletimSistemi,islemciTip,islemciNesli,ram,diskBoyut,diskTur,ekranBoyutu)
            else:
                UrunId=bilgiler
                urun_guncelle(fiyat,url, UrunId)

    else:
        modelId=model_ekle(markaId,ModelNo,ürünResim,ürünPuan)
        UrunId=urun_ekle(ürünAd,modelId,fiyat,url)
        urun_aciklama_ekle(UrunId,markaId,isletimSistemi,islemciTip,islemciNesli,ram,diskBoyut,diskTur,ekranBoyutu)
    
db.close()
```
