```python
ürün_bilgileri=[]
def ürün_bilgileri_bul(url):
    try:
        global ekran_boyutu,disk_türü,islemci_tipi,disk_kapasite,işletim_sistemi,model,ram,islemci_nesli,ssd,hdd
        istek1=requests.get(url,headers=header)
        soup1=BeautifulSoup(istek1.content,"lxml")
        urun_markasi=soup1.find("span",attrs={"class":"brand-name"}).text.strip().upper()
        model=soup1.find("h1",attrs={"class":"product-name best-price-trick"}).text.upper().strip()
        urun_puani=soup1.find("span",attrs={"class":"rating-star"}).text.strip()
        urunun_fotografi=soup1.img.get("src")
        urun_fiyati=soup1.find("span",attrs={"id":"offering-price"}).text.strip().replace("\n","").replace("\r","").replace("Adet","").replace("( )","").upper()
        aciklama=soup1.find("table",attrs={"class":"data-list tech-spec"})
        gt1=aciklama.find_all("tr")
        for i in gt1:
            gt2=i.find("th").get_text()
            if gt2.find("Ekran Boyutu")!=-1:
                ekran_boyutu=i.find("td").get_text().strip()
                if ekran_boyutu.find("inç")!=-1:
                    ekran_boyutu=ekran_boyutu.replace("inç","")
                if ekran_boyutu.find(",")!=-1:
                    ekran_boyutu=ekran_boyutu.replace(",",".")
            elif gt2.find("İşlemci Nesli")!=-1:
                islemci_nesli=i.find("td").get_text().upper().strip()
                if islemci_nesli.find("YOK")!=-1:
                    islemci_nesli="BELIRTILMEMIS"
            elif gt2.find("İşletim Sistemi")!=-1:
                işletim_sistemi=i.find("td").get_text().upper().strip()
                if işletim_sistemi.find("YOK")!=-1:
                    işletim_sistemi="FREEDOS"
                elif işletim_sistemi.find("WINDOWS")!=-1:
                    işletim_sistemi="WINDOWS"
                elif işletim_sistemi.find("DOS")!=-1:
                    işletim_sistemi="FREEDOS"
                elif işletim_sistemi.find("WIN")!=-1:
                    işletim_sistemi="WINDOWS"
                elif işletim_sistemi.find("CHROMEOS")!=-1:
                    işletim_sistemi="CHROMEOS"
                elif işletim_sistemi.find("MACOS")!=-1:
                    işletim_sistemi="MACOS"
            elif gt2.find("İşlemci Tipi")!=-1:
                islemci_tipi=i.find("td").get_text().upper().strip()
            elif gt2.find("Ram (Sistem Belleği)")!=-1:
                ram=i.find("td").get_text().strip()
            elif gt2.find("SSD Kapasitesi")!=-1:
                ssd=i.find("td").get_text().strip()
            elif gt2.find("Harddisk Kapasitesi")!=-1:
                hdd=i.find("td").get_text().strip()
        if ssd.find("Yok")!=-1:
            disk_kapasite=hdd
            disk_türü="HDD"
            print(disk_kapasite+ " "+disk_türü)
        elif hdd.find("Yok")!=-1:
            disk_kapasite=ssd
            disk_türü="SSD"
    except:
        pass
    else:
        print(urun_markasi,model,işletim_sistemi,islemci_tipi,islemci_nesli,ram,disk_kapasite,disk_türü,ekran_boyutu,urun_puani,urunun_fotografi,urun_fiyati,url)
        ürün_bilgileri.append([urun_markasi,model,işletim_sistemi,islemci_tipi,islemci_nesli,ram,disk_kapasite,disk_türü,ekran_boyutu,urun_puani,urunun_fotografi,urun_fiyati,url])


site_url="https://www.hepsiburada.com/laptop-notebook-dizustu-bilgisayarlar-c-98"
for k in range(2,80):
    try:
        istek=requests.get(site_url,headers=header,verify=False)
        soup=BeautifulSoup(istek.content,"lxml")        
        gt1=soup.find_all("li",attrs={"class":"productListContent-zAP0Y5msy8OHn5z7T_K_"})
        ürün_url=[]
        for i in gt1:
            ekle="https://www.hepsiburada.com"+i.find("a").get("href")
            ürün_url.append(ekle)
        for i in ürün_url:
            ürün_bilgileri_bul(i)
        site_url="https://www.hepsiburada.com/laptop-notebook-dizustu-bilgisayarlar-c-98?sayfa="+(k)
    except:
        pass
  
```
