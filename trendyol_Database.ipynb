```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import ast
import urllib3
urllib3.disable_warnings()
header={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36'}
```

```python
def ürün_bilgileri_bul(url):
    try:
        global marka,ekran_boyutu,disk_türü,işlemci,disk_kapasite,işletim_sistemi,model,ram,işlemci_nesli,hrd,model_no,ssd
        istek2=requests.get(url,headers=header,verify=False)
        soup2=BeautifulSoup(istek2.content,"lxml")
        ürün_ad=soup2.find("h3",attrs={"class":"detail-name"}).get_text().strip().upper()
        fiyat=soup2.find("span",attrs={"class":"prc-dsc"}).get_text()
        ürün_fiyat=fiyat[0:fiyat.find("TL")]
        bul=ürün_ad.find(" ")
        if bul!=-1:
            marka=ürün_ad[0:bul]
        aciklamalar=soup2.find_all("li",attrs={"class":"detail-attr-item"})
        for i in aciklamalar:
            gt=i.get_text()
            if  gt.find("İşlemci Tipi")!=-1:
                işlemci_tipi=i.find("b").get_text().upper().strip()
            elif gt.find("İşletim Sistemi")!=-1:
                işletim_sistemi=i.find("b").get_text().replace(" ","").upper()
                if işletim_sistemi.find("WINDOWS")!=-1:
                    işletim_sistemi="WINDOWS"
                elif işletim_sistemi.find("YOK")!=-1:
                    işletim_sistemi="YOK"
                elif işletim_sistemi.find("APPLE")==-1 and marka.find("APPLE")!=-1:
                    işletim_sistemi="APPLE "+işletim_sistemi
                elif işletim_sistemi.find("CHROMEOS")!=-1:
                    işletim_sistemi="CHROMEOS"
                elif işletim_sistemi.find("MACOS")!=-1:
                    işletim_sistemi="MACOS"
            elif gt.find("Ram (Sistem Belleği")!=-1 and gt.find("Ram (Sistem Belleği) Tipi")==-1:
                ram=i.find("b").get_text().strip()
            elif gt.find("Ekran Boyutu")!=-1:
                ekran_boyutu=i.find("b").get_text().strip()
                if ekran_boyutu.find("inç")!=-1:
                    ekran_boyutu=ekran_boyutu.replace("inç","")
                if ekran_boyutu.find(",")!=-1:
                    ekran_boyutu=ekran_boyutu.replace(",",".")
            elif gt.find("İşlemci Nesli")!=-1:
                işlemci_nesli=i.find("b").get_text().strip()
                if işlemci_nesli.find("Yok")!=-1:
                    işlemci_nesli="Belirtilmemiş"
                işlemci_nesli=işlemci_nesli.upper() 
            elif gt.find("İşlemci Tipi")!=-1:
                işlemci_tipi=i.find("b").get_text().upper().strip()
            elif gt.find("SSD Kapasitesi")!=-1:
                ssd=i.find("b").get_text().strip()
            elif gt.find("Hard Disk Kapasitesi")!=-1:
                hrd=i.find("b").get_text()
            elif gt.find("Kapasite")!=-1:
                disk_kapasite=i.find("b").get_text()
                if disk_kapasite.find("Yok")!=-1:
                    if ssd.find("Yok")==-1:
                        disk_kapasite=ssd
                    else:
                        disk_kapasite=hrd
        if ssd.find("Yok")==-1 and hrd.find("Yok")==-1:
            disk_türü="SSD-HDD"
        elif hrd.find("Yok")==-1:
            disk_türü="HDD"
        else:
            disk_türü="SSD"
    except:
        pass
    else:
        print(marka,ürün_ad,işletim_sistemi,işlemci_tipi,işlemci_nesli,ram,disk_kapasite,disk_türü,ekran_boyutu,ürün_fiyat,url)
        ürün_bilgileri.append([marka,ürün_ad,işletim_sistemi,işlemci_tipi,işlemci_nesli,ram,disk_kapasite,disk_türü,ekran_boyutu,ürün_fiyat,url])

    
site_url="https://www.trendyol.com/laptop-x-c103108"
ürün_bilgileri=[]
for k in range(2,80):
    try:
        istek=requests.get(site_url,headers=header,verify=False)
        soup=BeautifulSoup(istek.content,"lxml")
        gt1=soup.find_all("div",attrs={"class":"p-card-chldrn-cntnr card-border"})
        ürün_url=[]
        for i in range(0,len(gt1)):
            ürün_url.append("https://www.trendyol.com"+gt1[i].find("a").get("href"))
        for x in ürün_url:
            ürün_bilgileri_bul(x)
        site_url="https://www.trendyol.com/laptop-x-c103108?pi="+str(k)
    except:
        pass
```
